```python
import pandas as pd
import matplotlib.pyplot as plt

# GameStop  
gamestop_stock_data_extended = pd.DataFrame({
    "Date": ["2023-10-20", "2023-10-19", "2023-10-18", "2023-10-17", "2023-10-16",
             "2023-10-13", "2023-10-12", "2023-10-11", "2023-10-10", "2023-10-09"],
    "Open": [15.00, 15.25, 15.40, 15.30, 15.05, 14.80, 14.85, 15.00, 14.75, 14.65],
    "High": [15.30, 15.45, 15.60, 15.50, 15.20, 15.00, 15.10, 15.20, 15.00, 14.90],
    "Low": [14.90, 15.05, 15.20, 15.10, 14.90, 14.70, 14.80, 14.85, 14.60, 14.50],
    "Close": [15.20, 15.35, 15.50, 15.45, 15.10, 14.85, 14.95, 15.05, 14.80, 14.70]
})

#    
def make_graph(stock_data, title, price_column):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data['Date'], stock_data[price_column], label=f'{price_column} Price')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Stock Price ($)")
    plt.legend()
    plt.grid(True)
    plt.show()

# GameStop       
def make_multiple_gamestop_graphs(stock_data):
    make_graph(stock_data[['Date', 'Close']], "GameStop Stock Closing Price Over Time", 'Close')
    make_graph(stock_data[['Date', 'Open']], "GameStop Stock Opening Price Over Time", 'Open')
    make_graph(stock_data[['Date', 'High']], "GameStop Stock High Price Over Time", 'High')
    make_graph(stock_data[['Date', 'Low']], "GameStop Stock Low Price Over Time", 'Low')

#  
make_multiple_gamestop_graphs(gamestop_stock_data_extended)
```
